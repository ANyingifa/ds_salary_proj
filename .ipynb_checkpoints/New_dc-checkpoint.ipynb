```python
import pandas as pd
import numpy as np
import re 
```

```python
df = pd.read_csv("glassdoor_jobs.csv",index_col =0)
```

```python
df["Salary Estimate"].value_counts()
```

```python
df.drop(df[df["Salary Estimate"] == "-1"].index,inplace=True)
```

```python
df["Per_Hour"] = df["Salary Estimate"].apply(lambda x: 1 if "per hour" in x.lower() else 0)
```

```python
df["Employer_Provided"] = df["Salary Estimate"].apply(lambda x: 1 if "employer provided salary" in x.lower() else 0)
```

```python
def clean_salaries(text,fn="avg"):
    if (("per hour" in text.lower()) & ("employer provided" in text.lower())):
        cleaned_text = text.lower().split(":")[1].split("per hour")[0].strip()
    elif (("per hour" in text.lower()) & ("employer provided" not in text.lower())):
        cleaned_text = text.lower().split("per hour")[0].strip()
    elif (("per hour" not in text.lower()) & ("employer provided" in text.lower())):
        cleaned_text = text.split(":")[1].strip()
    else:
        cleaned_text = no_glass = text.split("(")[0].strip()
    no_currency = re.sub("[a-zA-Z$]","",cleaned_text)
    limits = [int(i.strip()) for i in no_currency.split("-") if i != ""]
    if fn == "min":
        return limits[0]
    elif fn == "max":
        return limits[1]
    else:
        return (limits[0]+limits[1])/2
        
    
        
    
        
    
```

```python
df[df["Rating"] == -1].iloc[:]["Company Name"].apply(lambda x:x.split("\r\n")[0])
```

```python
df["Min_Salary"] = df["Salary Estimate"].apply(lambda x: clean_salaries(x,fn="min"))
```

```python
df["Max_Salary"] = df["Salary Estimate"].apply(lambda x: clean_salaries(x,fn="max"))
```

```python
df["Average_Salary"] = df["Salary Estimate"].apply(lambda x: clean_salaries(x))
```

```python
df["Company Name"]= df["Company Name"].apply(lambda x:x.split("\r\n")[0])
```

```python
df["Job_State"] = df["Location"].apply(lambda x:x.strip()[-2:])
```

```python
df["Job_in_HQ"] = 1*(df["Location"] == df["Headquarters"])
```

```python
df["Age"] = df["Founded"].apply(lambda x:(2020 - x) if x!= -1 else x )
```

```python
import seaborn as sns
sns.set()
import matplotlib.pyplot as pt 
%matplotlib inline 
import plotly.express as px
```

```python
#Desired Skills
#Python
#R Studio
#Spark
#AWS
#Excel
```

```python
skill_set = ["Python","r studio","Spark","AWS","Excel"]
```

```python
job_skills = np.zeros(shape = (len(df),len(skill_set)),dtype="int")
```

```python
for i in range(len(df)):
    for p in range(len(skill_set)):
        if (skill_set[p].lower() in df.iloc[i]["Job Description"].lower()):
            job_skills[i,p] = 1
```

```python
job_skills_df = pd.DataFrame(job_skills,columns = skill_set,index = df.index)
```

```python
df[skill_set] = job_skills_df
```
